```python
### This section will be for Men's Swimming Team ###

import requests
from bs4 import BeautifulSoup
URL = ['https://www.brooklyncollegeathletics.com/sports/mens-swimming-and-diving/roster',
       'https://athletics.baruch.cuny.edu/sports/mens-swimming-and-diving/roster',
       'https://yorkathletics.com/sports/mens-swimming-and-diving/roster']
### iterate through every link and put it in the list)
page=[]
for item in URL:
    page.append(requests.get(item))
    
Parent_List_Height=[]
Parent_List_Height_CM=[]
thenamelist=[]

# Scrap Height of the player
for item in page:
    soup = BeautifulSoup(item.content,'html.parser')
    Menswimming = soup.find_all('span',class_='sidearm-roster-player-height')
    
    Child_List_Height=[]
    for i in Menswimming:
        Child_List_Height.append(i.get_text())    
    Parent_List_Height.append(Child_List_Height)
     
# Scrap name of the player      
for item in page: 
    soup = BeautifulSoup(item.content,'html.parser')
    Menswimmingname = soup.find_all('div',class_='sidearm-roster-player-name')
    for i in Menswimmingname:
        thenamelist.append(i.get_text().strip('\n').strip('C').strip('\n'))
        
### Trimming the duplication of height data ###
Parent_List_Height[0]=Parent_List_Height[0][:12]
Parent_List_Height[1]=Parent_List_Height[1][:23]        
Parent_List_Height[2]=Parent_List_Height[2][:10] 

### Converting Height to cm ###
for item in Parent_List_Height:
    for i in item:
        feet_cm = int(i[0:i.find('\'')])*30.48
        inc_cm = int(i[i.find('\'')+1:i.find('"')])*2.54
        total = total + (feet_cm+inc_cm)
        Parent_List_Height_CM.append(feet_cm+inc_cm)
         
#print(Parent_List_Height_CM)
#print(thenamelist)

### Creating a Dataframe ###
import pandas as pd
menswimmersdf=pd.DataFrame()
menswimmersdf['Name']                           =thenamelist
menswimmersdf['Height']                         =Parent_List_Height_CM

menswimmersdf
```

```python
mean = menswimmersdf['Height'].mean()
highest_menswimmersdf = menswimmersdf.nlargest(5, 'Height')
smallest_menswimmersdf = menswimmersdf.nsmallest(5, ['Height'])
print ("Average Men Swimmers Team {} cm".format(mean))
print("\n5 highest Men Swimmers height\n",highest_menswimmersdf)
print("\n5 smallest Men Swimmers height\n",smallest_menswimmersdf)
```
