```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

downloading contents of the webpage

```python
url = 'https://chartmasters.org/most-streamed-artists-ever-on-spotify/'
data = requests.get(url).text

# soup object
soup = BeautifulSoup(data, 'html.parser')
```

```python
for data in soup.find_all('table'):
    print(data)
```

creating dataframe and looping through each row(tr) in the table

```python
streaming_data = soup.find('table')
columns = ['rank', 'artist_name', 'lead_streams', 'featured_streams', 'tracks', '1B+', '100M+', '10M+', '1M+', 'last_update']
streaming_data_dataframe = pd.DataFrame(columns = columns)

for row in streaming_data.tbody.find_all('tr'):
    row_data = row.find_all('td')

    if row_data != []:
        rank = row_data[0].text.strip()
        artist_name = row_data[1].text.strip()
        lead_streams = row_data[2].text.strip()
        featured_stream = row_data[3].text.strip()
        tracks = row_data[4].text.strip()
        b = row_data[5].text.strip()
        m100 = row_data[6].text.strip()
        m10 = row_data[7].text.strip()
        m1 = row_data[8].text.strip()
        last_update = row_data[9].text.strip()

        streaming_data_dataframe = streaming_data_dataframe.append({
            'rank' : rank,
            'artist_name' : artist_name,
            'lead_streams' : lead_streams,
            'featured_streams' : featured_stream ,
            'tracks' : tracks,
            '1B+' : b,
            '100M+' : m100,
            '10M+' : m10,
            '1M+' : m1,
            'last_update' : last_update}, ignore_index = True)
```

Convert numeric columns to int

```python
def to_int(column: str) -> None:
    """
    removes commas in passed column and converts column type to int
    :param column
    :return:
    """
    streaming_data_dataframe[column] = streaming_data_dataframe[column].str.replace(',','')
    streaming_data_dataframe[column] = streaming_data_dataframe[column].astype(int)

to_int('lead_streams')
to_int('featured_streams')
to_int('tracks')
to_int('1B+')
to_int('100M+')
to_int('10M+')
to_int('1M+')

streaming_data_dataframe.info()
```

saving dataframe as a csv file

```python
streaming_data_dataframe.to_csv("most_streamed_artist_spotify.csv", index = False)
```
